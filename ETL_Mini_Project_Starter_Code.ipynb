### Create the Category and Subcategory DataFrames
---
**Create a Category DataFrame that has the following columns:**
- A "category_id" column that is numbered sequential form 1 to the length of the number of unique categories.
- A "category" column that has only the categories.

Export the DataFrame as a `category.csv` CSV file.

**Create a SubCategory DataFrame that has the following columns:**
- A "subcategory_id" column that is numbered sequential form 1 to the length of the number of unique subcategories.
- A "subcategory" column that has only the subcategories. 

Export the DataFrame as a `subcategory.csv` CSV file.

```python
# Get the crowdfunding_info_df columns.
print(crowdfunding_info_df.columns)
```

```python


# Define the categories and subcategories
categories = ['food', 'music', 'technology', 'theater', 'film & video', 'publishing', 'games',
 'photography', 'journalism']

subcategories = [
    'food trucks', 'rock', 'web', 'plays', 'documentary', 'electric music', 'drama',
    'indie rock', 'wearables', 'nonfiction', 'animation', 'video games', 'shorts',
    'fiction', 'photography books', 'radio & podcasts', 'metal', 'jazz',
    'translations', 'television', 'mobile games', 'world music',
    'science fiction', 'audio'
]

# Ensure that the lengths of categories and subcategories match
if len(categories) != len(subcategories):
    print("The number of categories and subcategories do not match!")
else:
    # Assign categories and subcategories to the 'crowdfunding_info_df' DataFrame
    crowdfunding_info_df['category'] = categories * (len(crowdfunding_info_df) // len(categories)) + categories[:len(crowdfunding_info_df) % len(categories)]
    crowdfunding_info_df['subcategory'] = subcategories * (len(crowdfunding_info_df) // len(subcategories)) + subcategories[:len(crowdfunding_info_df) % len(subcategories)]

# Now the 'category' and 'subcategory' columns will have the appropriate values

# Split the 'category & sub-category' column into two columns: 'category' and 'subcategory'
if 'category & sub-category' in crowdfunding_info_df.columns:
    crowdfunding_info_df[['category', 'subcategory']] = crowdfunding_info_df['category & sub-category'].str.split('/', expand=True)
    
    # Optionally, you can fill any NaN values with a default value if needed
    crowdfunding_info_df['category'].fillna('Unknown', inplace=True)
    crowdfunding_info_df['subcategory'].fillna('Unknown', inplace=True)

# Display the updated DataFrame with the split columns
print(crowdfunding_info_df[['category', 'subcategory']].head())
```

```python
# Get the unique categories and subcategories in separate lists.


print(categories)
print(subcategories)
```

```python
# Get the number of distinct values in the categories and subcategories lists.
print(len(categories))
print(len(subcategories))
```

```python
# Create numpy arrays from 1-9 for the categories and 1-24 for the subcategories.
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1, 25)

print(category_ids)
print(subcategory_ids)
```

```python
# Use a list comprehension to add "cat" to each category_id. 
cat_ids = [f"cat{num}" for num in category_ids]

scat_ids = [f"subcat{num}" for num in subcategory_ids]
# Use a list comprehension to add "subcat" to each subcategory_id.    

    
print(cat_ids)
print(scat_ids)
```

```python
# Create a category DataFrame with the category_id array as the category_id and categories list as the category name.
import pandas as pd

# Create category DataFrame
category_df = pd.DataFrame({
    'category_id': cat_ids,  # category_ids with "cat" prefix
    'category': categories    # categories list
})

# Create subcategory DataFrame
subcategory_df = pd.DataFrame({
    'subcategory_id': scat_ids,  # subcategory_ids with "subcat" prefix
    'subcategory': subcategories  # subcategories list
})

# Display both DataFrames
print(category_df)
print(subcategory_df)
```

### Campaign DataFrame
----
**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The “company_name” column.
- The "blurb" column is renamed as "description."
- The "goal" column.
- The "goal" column is converted to a `float` datatype.
- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.
- The "currency" column.
- The "launched_at" column is renamed as "launch_date" and converted to a datetime format. 
- The "deadline" column is renamed as "end_date" and converted to a datetime format.
- The "category_id" with the unique number matching the “category_id” from the category DataFrame. 
- The "subcategory_id" with the unique number matching the “subcategory_id” from the subcategory DataFrame.
- And, create a column that contains the unique four-digit contact ID number from the `contact.xlsx` file.
 

Then export the DataFrame as a `campaign.csv` CSV file.


```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the columns 'blurb', 'launched_at', and 'deadline'
crowdfunding_info_df.rename(columns={
    'blurb': 'description',  # Rename 'blurb' to 'description'
    'launched_at': 'launch_date',  # Rename 'launched_at' to 'launch_date'
    'deadline': 'end_date'  # Rename 'deadline' to 'end_date'
}, inplace=True)

# Display the updated DataFrame
print(crowdfunding_info_df.head())
```

```python
# Convert the goal and pledged columns to a `float` data type.
crowdfunding_info_df['goal'] = crowdfunding_info_df['goal'].astype(float)
crowdfunding_info_df['pledged'] = crowdfunding_info_df['pledged'].astype(float)
print(crowdfunding_info_df.head())
```

```python
# Check the datatypes
print(crowdfunding_info_df.dtypes)
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt

# Convert 'launched_date' and 'end_date' from Unix timestamp to datetime format
crowdfunding_info_df['launch_date'] = pd.to_datetime(crowdfunding_info_df['launch_date'], unit='s')
crowdfunding_info_df['end_date'] = pd.to_datetime(crowdfunding_info_df['end_date'], unit='s')

print(crowdfunding_info_df[['launch_date', 'end_date']].head())
```

```python

# Merge campaign_df with category_df on the "category" column
campaign_merged_df = pd.merge(campaign_df, category_df, how='left', left_on='category', right_on='category')

# Merge the resulting DataFrame with subcategory_df on the "subcategory" column
campaign_merged_df = pd.merge(campaign_merged_df, subcategory_df, how='left', left_on='subcategory', right_on='subcategory')

# Display the last 10 rows of the merged DataFrame
print(campaign_merged_df.tail(10))
```

```python
# Drop unwanted columns
columns_to_drop = ['category_id', 'subcategory_id']
campaign_cleaned_df = campaign_merged_df.drop(columns=columns_to_drop)

# Display the cleaned DataFrame
print(campaign_cleaned_df.head())
```

```python
# Export the DataFrame as a CSV file. 
campaign_cleaned_df.to_csv("Resources/campaign.csv", index=False)
```

### Option 1: Use Pandas to create the contacts DataFrame.

```python
print(contact_info_df.columns.tolist())
```

```python
import json
# Load the Excel file
file_path = "Resources/contacts.xlsx"
contact_info_df = pd.read_excel(file_path, skiprows=2)  # Skip the first two rows

# Rename the first column to 'contact_info' (if needed)
contact_info_df.columns = ['contact_info']

# Ensure the correct column name
if 'contact_info' in contact_info_df.columns:
    dict_values = []

    # Iterate over rows and convert each JSON string into a dictionary
    for index, row in contact_info_df.iterrows():
        try:
            data = json.loads(row['contact_info'])  # Convert JSON string to dictionary
            name_parts = data.get('name', '').split()
            first_name = name_parts[0] if len(name_parts) > 0 else ''
            last_name = name_parts[1] if len(name_parts) > 1 else ''

            row_values = [
                data.get('contact_id'),
                first_name,
                last_name,
                data.get('email')
            ]
            dict_values.append(row_values)
        except (json.JSONDecodeError, TypeError, AttributeError) as e:
            print(f"Skipping row {index} due to error: {e}")

    # Create a DataFrame from the extracted values
    contact_info_cleaned = pd.DataFrame(dict_values, columns=['contact_id', 'first_name', 'last_name', 'email'])

    # Display the cleaned DataFrame
    print(contact_info_cleaned.head())

else:
    print("Error: 'contact_info' column not found in DataFrame.")


```

```python
# Create a contact_info DataFrame and add each list of values, i.e., each row 
# to the 'contact_id', 'name', 'email' columns.
contact_info_df = pd.DataFrame(dict_values, columns=['contact_id', 'first_name', 'last_name', 'email'])

# Display the DataFrame
contact_info_df.head()



```
